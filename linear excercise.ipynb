```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("canada_per_capita_income.csv")
df
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[["year"]],df.capital_income)
```

```python
%matplotlib inline
plt.xlabel("year ")
plt.ylabel("capital_income(US$)")
plt.scatter(df.year,df.capital_income,color='red',marker='+')
```

```python
reg.predict([[2020]])
```

```python
d=pd.read_csv("years.csv")
d.head(3)
```

```python
p = reg.predict(d)
```

```python
d['income'] = p
```

```python
d.to_csv("predic2.csv")
```
